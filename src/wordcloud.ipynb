```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
def get_topic_text(filename, num):
    topic_df = pd.read_csv(filename)
    words = topic_df["words"]
    weights = topic_df[str(num)]

    text = ""
    for i in range(len(words)):
        word = str(words[i])
        repeats = float(weights[i]) * 1e7
        for j in range(int(repeats)):
            text += (word + " ")
    
    return text
```

```python
def get_word_cloud(filename, num):
    text = get_topic_text(filename, num)

    # Generate word cloud
    word_cloud = WordCloud(
        width=3500,
        height=3000,
        random_state=1,
        background_color="white",
        collocations=False,
        prefer_horizontal = 0.6
        ).generate(text)

    # Display the generated Word Cloud
    plt.imshow(word_cloud)
    plt.axis("off")
    plt.show()
```

```python
get_word_cloud("../data/processed_LDA_files/blog.csv", 1)
```
